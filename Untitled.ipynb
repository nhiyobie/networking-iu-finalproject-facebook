```python
#WEEK 9 work -- might be useful

import networkx as nx
import scipy.stats as stats

# load the cond-mat network
condmat_net = nx.read_gml("cond-mat_01.gml")

condmat_net = nx.Graph(condmat_net)
#print(type(condmat_net).__name__)

# measure and report the size of the largest robins_alexander_clustering in the graph.

average_clustering = nx.average_clustering(condmat_net)
condmat_degree_sequence = list(dict(nx.degree(condmat_net)).values())


# We need to generate enough null graphs to make a reasonable assessment of the null distribution
model_average_clustering = []
for i in range(50):
    # Use the configuration model to create a null graph
    null_graph = nx.configuration_model(condmat_degree_sequence)
 
    # remove self-loops and parallel edges 
    null_graph = nx.Graph(null_graph) # removes parallel edges
    null_graph.remove_edges_from(null_graph.selfloop_edges()) # removes self-loops
    
    # Calculate the model's max size and add it to the list
    model_average_clustering.append(nx.average_clustering(null_graph))

   
zscores = stats.zscore([average_clustering] + model_average_clustering)
zscores = zscores[1:50]

%matplotlib inline
import matplotlib.pyplot as plt

# Use the histogram function to plot the distribution of assortativity coefficients

plt.hist(zscores, bins=10)

# axvline produces a vertical line at the designated x position on the graph
plt.axvline(average_clustering, lw=2, color="red") #lw = 2 makes the line thicker
plt.show()
```
