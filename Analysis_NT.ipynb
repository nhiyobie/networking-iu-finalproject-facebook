```python
import networkx as nx
import pandas as pd
```

```python
#read facebook dataset
#edge file
edges_list_df=pd.read_csv('musae_facebook_edges.csv')
node_attributes=pd.read_csv('musae_facebook_target.csv')
```

```python
edges_list_df
```

```python
node_attributes
```

```python
G = nx.Graph()
```

```python
 G.add_edges_from(edges_list_df.values.tolist())
```

```python
G
```

```python
len(G)
```

```python
print(nx.info(G))
```

```python
node_dict=node_attributes.to_dict('index')
```

```python
nx.set_node_attributes(G,node_dict)
```

```python
G.nodes[0]['page_type']
```

```python
nx.write_gexf(G,'facebook.gexf')
```

```python
largest_component=max(nx.connected_component_subgraphs(G), key=len)
```

```python
print(nx.info(largest_component))
```

```python
G.remove_edges_from(nx.selfloop_edges(G))
```

```python
print(nx.info(G))
```

```python
dict(G.nodes.data())
```

```python
values=dict(G.nodes.data()).values()
```

```python
values_df=pd.DataFrame(values)
```

```python
values_df['page_type'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
values_df['page_type'].value_counts().plot.pie(subplots=True, figsize=(15,12),autopct='%1.1f%%', legend=False, cmap='Dark2')
plt.title('Percentage of Page Type In Page-Page Network Graph')
plt.ylabel('')
plt.rcParams['font.size'] = 16.0
```

```python
degree_graph_count=pd.DataFrame(G.degree(), columns=['node_id','degree'])
```

```python
degree_graph_count_sorted=degree_graph_count.sort_values(by=['degree'],ascending=False)
```

```python
degree_graph_count.iloc[:,1:2].hist(bins=250)
plt.xlim(0,150)
```
